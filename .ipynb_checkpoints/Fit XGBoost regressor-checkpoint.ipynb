```python
from fastai.imports import *
import dtale
import klib
import seaborn as sns
from numpy import random
from sklearn.model_selection import train_test_split as t
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from xgboost import XGBRegressor
np.set_printoptions(linewidth=130)
```

```python
def clean_data(df):
    klib.data_cleaning(df)
    klib.clean_column_names(df)
    df_train=klib.convert_datatypes(df)
    klib.mv_col_handling(df)
```

```python
def proc_data(df):
    df['pop']=np.log1p(df['pop'])
    df.drop(['date','id'],axis=1,inplace=True)
    df['city'] = pd.Categorical(df.city)
    df['capacity'] = pd.Categorical(df.capacity)
    df['shop'] = pd.Categorical(df.shop)
    df['container'] = pd.Categorical(df.container)
    df['brand'] = pd.Categorical(df.brand)
```

```python
clean_data(df)
proc_data(df)
```

```python
df.shape
```

```python
cats=['city','capacity','shop','container','brand']
conts=['lat','long','pop','price']
dep='quantity'
```

```python
fig,axs = plt.subplots(1,2, figsize=(11,5))
sns.barplot(data=df, y=dep, x="city", ax=axs[0]).set(title="sales")
sns.countplot(data=df, x="city", ax=axs[1]).set(title="Histogram");
```

```python
random.seed(42)
trn_df,val_df = train_test_split(df, random_state=42, test_size=0.2)
trn_df[cats] = trn_df[cats].apply(lambda x: x.cat.codes)
val_df[cats] = val_df[cats].apply(lambda x: x.cat.codes)
```

```python
val_df.head()
```

```python
df_pop = trn_df[trn_df['pop']>0]
fig,axs = plt.subplots(1,2, figsize=(11,5))
sns.scatterplot(data=df_pop, x=dep, y="pop", ax=axs[0])
sns.kdeplot(data=df_pop, x="pop", ax=axs[1]);
```

```python
def scale_data(train,test):
    sc= StandardScaler()
    Y_train=train['quantity']
    Y_test=test['quantity']    
    X_train=train.drop('quantity',axis=1)
    X_train_std=sc.fit_transform(X_train)
    X_test=test.drop('quantity',axis=1)
    X_test_std=sc.transform(X_test)
    return X_train_std,Y_train,X_test_std,Y_test
```

```python
X_train_std,Y_train,X_test_std,Y_test=scale_data(trn_df,val_df)
```

```python
Y_train.shape
```

```python
X_train_std.shape
```

```python
regressor = XGBRegressor()
regressor.fit(X_train_std,Y_train)
```

```python
training_data_prediction = regressor.predict(X_train_std)
```

```python
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('R Squared value = ', r2_train)
```

```python
test_data_prediction = regressor.predict(X_test_std)
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('R Squared value = ', r2_test)
```

```python
print(mean_absolute_error(Y_test,test_data_prediction))
```

```python
regressor = XGBRegressor(max_depth=3,subsample=1,n_estimators=1000,
                         learning_rate=0.05,min_child_weight=1,random_state=5,reg_alpha=0,reg_lambda=1)
regressor.fit(X_train_std,Y_train)
training_data_prediction = regressor.predict(X_train_std)
r2_train = metrics.r2_score(Y_train, training_data_prediction)
print('Train R Squared value = ', r2_train)
test_data_prediction = regressor.predict(X_test_std)
r2_test = metrics.r2_score(Y_test, test_data_prediction)
print('Test R Squared value = ', r2_test)
```
